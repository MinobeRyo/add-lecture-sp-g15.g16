```python
import pandas as pd
df = pd.read_csv('stats_bat.csv', encoding='cp932')
```

```python
df = df.drop(columns=['h', 'gs', 'r', 'h', 'h1b', 'h2b', 'h3b', 'so', 'tb', 'sh', 'sf', 'sb', 'cs', 'e', 'gidp', 'pb'])
```

```python
df_filtered = df[(df['team_league_name'] == 'パ・リーグ') & (df['pa'] >= 200)]
print(df_filtered[['player_name', 'hr', 'rbi', 'bb', 'hbp', 'ibb']])
print(f"データの件数: {len(df_filtered)} 件")
```

```python
df_filtered = df[(df['team_league_name'] == 'パ・リーグ') & (df['pa'] >= 200)]
df_sorted = df_filtered.sort_values(by=['player_id', 'season_year'], ascending=[True, True])
list_columns = ['player_id', 'player_name', 'season_year', 'team_name', 'pa', 'ab', 'hr', 'rbi', 'bb', 'hbp', 'ibb']
df_sorted = df_sorted[list_columns]

df = df[df['hr'] >= 10].copy()

df['四死球'] = df['bb'] + df['hbp']

df['本塁打に対する四死球比率'] = df['四死球'] / df['hr']

#本塁打に対して四死球が少ない選手を表示

print(df.sort_values(by='本塁打に対する四死球比率')[['player_name', 'hr', '四死球', '本塁打に対する四死球比率']])
```
