```python
import pandas as pd
import unicodedata

df = pd.read_csv('stats_bat.csv', encoding='cp932')

df_target = df[
    (df['team_league_name'] == 'セ・リーグ') & 
    (df['pa'] >= 200)
].copy()

df_target['hr_rate'] = df_target['hr'] / df_target['pa']
df_target['bb_hbp_rate'] = (df_target['bb'] + df_target['hbp']) / df_target['pa']


x = df_target['bb_hbp_rate']
y = df_target['hr_rate']

x_mean = x.mean()
y_mean = y.mean()

numerator = ((x - x_mean) * (y - y_mean)).sum()
denominator = ((x - x_mean) ** 2).sum()
slope = numerator / denominator

intercept = y_mean - (slope * x_mean)

df_target['predicted_hr_rate'] = (slope * x) + intercept


df_target['residual'] = df_target['hr_rate'] - df_target['predicted_hr_rate']
df_target['outlier_score'] = df_target['residual'].abs()

outliers = df_target.sort_values('outlier_score', ascending=False).head(10)

print("=== ホームラン率と四死球率の関係における外れ値選手 ===")

display(outliers.reset_index(drop=True))
```
